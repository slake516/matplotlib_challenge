```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "Pymaceuticals/mouse_drug_data.csv"
clinicaltrial_data_to_load = "Pymaceuticals/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinicaltrial_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(clinical_data,mouse_data, how="outer", on="Mouse ID")
combined_df.head()
```

```python
# Reference: github - arlennav
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
displayColumns=[
      'Drug',
      'Timepoint',
      'Tumor Volume (mm3)']

tumor_mean=pd.DataFrame(combined_df.groupby(["Drug","Timepoint"],as_index = False).mean())[displayColumns]
tumor_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
clinical_tumor_sem=pd.DataFrame(combined_df.groupby(["Drug","Timepoint"])[['Tumor Volume (mm3)']].sem())
clinical_tumor_sem.reset_index(inplace=True)
clinical_tumor_sem.head()
```

```python

# Minor Data Munging to Re-Format the Data Frames
pivot_table=tumor_mean.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
pivot_table
```

```python
# Generate the Plot (with Error Bars)
listLegend=['Capomulin','Infubinol','Ketapril','Placebo']

plt.errorbar(pivot_table.index,pivot_table['Capomulin'], \
             yerr=clinical_tumor_sem[clinical_tumor_sem['Drug']=='Capomulin']['Tumor Volume (mm3)'],fmt="ro--",linewidth=0.5,label='Capomulin')
plt.errorbar(pivot_table.index,pivot_table['Infubinol'],\
             yerr=clinical_tumor_sem[clinical_tumor_sem['Drug']=='Infubinol']['Tumor Volume (mm3)'], fmt="b^--",linewidth=0.5,label='Infubinol')
plt.errorbar(pivot_table.index,pivot_table['Ketapril'],\
             yerr=clinical_tumor_sem[clinical_tumor_sem['Drug']=='Ketapril']['Tumor Volume (mm3)'], fmt="gs--",linewidth=0.5, label='Ketapril')
plt.errorbar(pivot_table.index,pivot_table['Placebo'],\
             yerr=clinical_tumor_sem[clinical_tumor_sem['Drug']=='Placebo']['Tumor Volume (mm3)'], fmt="kd--",linewidth=0.5, label='Placebo')


# Create a legend for our chart
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc=2)
# Show the chart
plt.grid()
# Save the Figure
plt.savefig("Tumor Response to Treatment.png")
```

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_mean=pd.DataFrame(combined_df.groupby(["Drug","Timepoint"]).mean())[['Metastatic Sites']]
met_mean.head()
```

```python

# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sem=pd.DataFrame(combined_df.groupby(["Drug","Timepoint"])[['Metastatic Sites']].sem())
met_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_mean.reset_index(inplace=True)
met_pivot=met_mean.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
met_pivot
```

```python
# Generate the Plot (with Error Bars)
listLegend=['Capomulin','Infubinol','Ketapril','Placebo']

plt.errorbar(met_pivot.index,met_pivot['Capomulin'], \
             yerr=met_sem.loc['Capomulin']['Metastatic Sites'],fmt="ro--",linewidth=0.5, label='Capomulin')
plt.errorbar(met_pivot.index,met_pivot['Infubinol'],\
             yerr=met_sem.loc['Infubinol']['Metastatic Sites'], fmt="b^--",linewidth=0.5, label='Infubinol')
plt.errorbar(met_pivot.index,met_pivot['Ketapril'],\
             yerr=met_sem.loc['Ketapril']['Metastatic Sites'], fmt="gs--",linewidth=0.5, label='Ketapril')
plt.errorbar(met_pivot.index,met_pivot['Placebo'],\
             yerr=met_sem.loc['Placebo']['Metastatic Sites'], fmt="kd--",linewidth=0.5, label='Placebo')


# Create a legend for our chart
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.legend(loc="best")
# Show the chart
plt.grid()
# Save the Figure
plt.savefig("Metastatic Spread During Treatment.png")
```

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
displayColumns=[
      'Drug',
      'Timepoint',
      'Mouse ID']

df_mice_count=pd.DataFrame(combined_df.groupby(["Drug","Timepoint"],as_index = False).count())[displayColumns].rename(columns={'Mouse ID': 'Mouse Count'})
df_mice_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_pivot=df_mice_count.pivot(index='Timepoint', columns='Drug', values='Mouse Count')
mice_pivot
```

```python
# Generate the Plot (Accounting for percentages)
listLegend=['Capomulin','Infubinol','Ketapril','Placebo']

plt.plot(mice_pivot.index,100 * mice_pivot['Capomulin'] / max(mice_pivot['Capomulin']),"ro--",linewidth=0.5,\
        label='Capomulin')
plt.plot(mice_pivot.index,100 * mice_pivot['Infubinol'] / max(mice_pivot['Infubinol']), "b^--",linewidth=0.5,\
        label='Infubinol')
plt.plot(mice_pivot.index,100 * mice_pivot['Ketapril'] / max(mice_pivot['Ketapril']), "gs--",linewidth=0.5,\
        label='Ketapril')
plt.plot(mice_pivot.index,100 * mice_pivot['Placebo'] / max(mice_pivot['Placebo']), "kd--",linewidth=0.5,\
        label='Placebo')

# Create a legend for our chart
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend(loc="best")

plt.ylim(31,103)
plt.grid()
# Save the Figure
plt.savefig("Survival During Treatment.png")
```

```python
# Calculate the percent changes for each drug
tumor_change = ((pivot_table.loc[45, :] - pivot_table.loc[0, :])/pivot_table.loc[0, :] * 100)[['Capomulin','Infubinol','Ketapril','Placebo']]
# Display the data to confirm
tumor_change
```

```python
# Store all Relevant Percent Changes into a Tuple
# Splice the data between passing and failing drugs
plt.title('Tumor Volume Change over 45 Day Treatment', fontdict = {'fontsize': 12, 'fontweight': 'bold'})
plt.ylabel('% Tumor Volume Change')
plt.axhline(y=0, color = 'black') #adds a horizontal line at zero
#drug names as labels
xlabels = tumor_change.index
plt.xticks(np.arange(len(xlabels)), xlabels)

#graphs % change of tumor as bar graph
change_plt=plt.bar(np.arange(4), 
        tumor_change,
        color = ['red' if tumor_change[r] > 0 else 'green' for r in np.arange(len(xlabels))]
       )
# Orient widths. Add labels, tick marks, etc. 


# Use functions to label the percentages of changes
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        if height > 0:
            label_position = 2
        else:
            label_position = -8
        plt.text(rect.get_x() + rect.get_width()/2., label_position,
                '%d' % int(height)+'%',color='white',
                ha='center', va='bottom')
                
# Call functions to implement the function calls
autolabel(change_plt)

# Save the Figure
# Show the Figure
plt.grid()
#plt.show()
plt.savefig("Tumor Volume Change over 45 Day Treatment.png")
```
